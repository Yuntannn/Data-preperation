# Check any missing data in the dataset

Use isnull( ) and isna( ) to check if there is any missing or empty data in the dataset, one of the method of data cleaning

```python
null_data = data.isnull().sum().sum()
nan_data = data.isna().sum().sum()


print('amount of null data: %s'%(null_data))
print('amount of nan data: %s'%(nan_data))
```

# Check any wrong data

We compare the value name between the dataset in the csv file and kaggle to check for wrong values

```python
for i in range(0,data.shape[1]):
    print(data[data.columns[i]].value_counts())
```

Convert the data from string to int (data transfer)to help us to find any relationship
we also extract the numerical data using to_csv( )under the dataset folder

```python
new_data = data

for i in range(0,data.shape[1]):
    new_data[new_data.columns[i]] = pd.factorize(new_data[new_data.columns[i]])[0]
new_data.to_csv ('dataset/newmushrooms.csv', index = False, header=True)
```

# Visualiazation

Now we come down to the visualization part, we use seaborn to plot all 3 graphs

The first bar graph is for determine the relationship between column odor and class(edible status)
Class 0 is stand for poisonous and Class 0 is edible, Odor 1-8 stands for different odors of the mushroom

catplot()plot visualizations, within the function kind = 'count' means besides y variable odor, the x variable will be the number of data appearence of ordor. And hue = 'class' will divide the ordor count data into edible and poisonous

```python
sns.catplot(y="odor", hue="class", kind="count",
            palette="rocket",edgecolor=".1",
            data=new_data);
```

The second and third visualization is to verify absoluteness of relationship between mushroom ordor and edible. We determine the relationship under different gill-size, the result is shown between 2 gill-sizes, the relationship is constand

for visualiazation 2 and 3, we use violin graph by using kind = 'violin' and set the split to be true to join the edible and posion data on different side of same y position

```python
sns.catplot(x="gill-size", y="odor", hue="class",
            kind="violin", inner="stick", split=True,
            palette="rocket", data=new_data);
```

However when we determine the relationship under bruises existance, the result is shown between different bruises existance status, the relationship between ordor and edible is completely different

```python
sns.catplot(x="bruises", y="odor", hue="class",
            kind="violin", inner="stick", split=True,
            palette="rocket", data=new_data);
```
